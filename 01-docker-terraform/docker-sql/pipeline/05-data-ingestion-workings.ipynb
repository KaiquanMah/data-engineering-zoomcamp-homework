### Handling Data Types


````python
dtype = {
    "VendorID": "Int64",
    "passenger_count": "Int64",
    "trip_distance": "float64",
    "RatecodeID": "Int64",
    "store_and_fwd_flag": "string",
    "PULocationID": "Int64",
    "DOLocationID": "Int64",
    "payment_type": "Int64",
    "fare_amount": "float64",
    "extra": "float64",
    "mta_tax": "float64",
    "tip_amount": "float64",
    "tolls_amount": "float64",
    "improvement_surcharge": "float64",
    "total_amount": "float64",
    "congestion_surcharge": "float64"
}

parse_dates = [
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime"
]

df = pd.read_csv(
    prefix + 'yellow_tripdata_2021-01.csv.gz',
    nrows=100,
    dtype=dtype,
    parse_dates=parse_dates
)
````

````python
# create table from DF schema
# `head(n=0)` makes sure we only create the table, we don't add any data yet.
df.head(n=0).to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
````

## Ingesting Data in Chunks

````python
url = 'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/yellow/'  + 'yellow_tripdata_2021-01.csv.gz'
url
````

````python
# We don't want to insert all the data at once. Let's do it in batches and use an iterator for that:
df_iter = pd.read_csv(
    url,
    dtype = dtype,
    parse_dates = parse_dates,
    iterator=True,
    chunksize=100000
)

counter = 1
# Iterate Over Chunks
for df_chunk in df_iter:
    print(f"Chunk {counter}: {len(df_chunk)}")
    counter+=1
````

````python
# recreate iterator
df_iter = pd.read_csv(
    url,
    dtype = dtype,
    parse_dates = parse_dates,
    iterator=True,
    chunksize=100000
)

# then look at the df chunks
next(df_iter)
````

````python
### Inserting Data
# df_chunk.to_sql(name='yellow_taxi_data', con=engine, if_exists='append')

### Approach 1
### Complete Ingestion Loop
first = True

for df_chunk in df_iter:

    if first:
        # Create table schema (no data)
        df_chunk.head(0).to_sql(
            name="yellow_taxi_data",
            con=engine,
            if_exists="replace"
        )
        first = False
        print("Table created")

    # Insert chunk
    df_chunk.to_sql(
        name="yellow_taxi_data",
        con=engine,
        if_exists="append"
    )

    print("Inserted:", len(df_chunk))
````

````python
### Approach 2
### Alternative Approach (Without First Flag)
first_chunk = next(df_iter)

first_chunk.head(0).to_sql(
    name="yellow_taxi_data",
    con=engine,
    if_exists="replace"
)

print("Table created")

first_chunk.to_sql(
    name="yellow_taxi_data",
    con=engine,
    if_exists="append"
)

print("Inserted first chunk:", len(first_chunk))

for df_chunk in df_iter:
    df_chunk.to_sql(
        name="yellow_taxi_data",
        con=engine,
        if_exists="append"
    )
    print("Inserted chunk:", len(df_chunk))
````

````python
from tqdm.auto import tqdm
````

````python
### Approach 1 with tqdm
# need to recreate the iterator
# because we 'exhausted' it earlier during counting
df_iter = pd.read_csv(
    url,
    dtype = dtype,
    parse_dates = parse_dates,
    iterator=True,
    chunksize=100000
)

first = True

counter = 1
for df_chunk in tqdm(df_iter):

    if first:
        # Create table schema (no data)
        df_chunk.head(0).to_sql(
            name="yellow_taxi_data",
            con=engine,
            if_exists="replace"
        )
        first = False
        print("Table created")

    # Insert chunk
    df_chunk.to_sql(
        name="yellow_taxi_data",
        con=engine,
        if_exists="append"
    )

    print(f"Inserted Chunk {counter}: {len(df_chunk)}")
    counter+=1
````
